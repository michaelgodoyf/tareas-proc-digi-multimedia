## 2. Define en python vectores y funciones
<p align="justify">
Debes generar vectores que representan intervalos de tiempo o de una variable.
Debes ingresar las funciones matemáticas y evaluarlas en los vectores generados.

$$
f(x) =  e^{-x^2}
$$

$$
f(z) =  ysin(x)

<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generar vector de valores de x en el intervalo
valor_x = np.linspace(-4, 4, 200)
valor_y= np.linspace(-4, 4, 200)

# Definir la función e^(-x^2)
def funcion_exponencial(x):
    return np.exp(-x**2)

# Definir la función f(z) = y * sin(x)
def funcion_seno(x, y):
    return y * np.sin(x)

# Evaluar la función exponencial en el intervalo
valores_exponencial = funcion_exponencial(valor_x)

# Visualizar los resultados de la función exponencial
plt.figure(figsize=(5,5))
plt.plot(valor_x, valores_exponencial, label='$e^{-x^2}$')
plt.xlabel('x')
plt.ylabel('$e^{-x^2}$')
plt.title('Función $e^{-x^2}$ evaluada en el intervalo [-4, 4]')
plt.legend()
plt.grid(True)
plt.show()

# Crear una cuadrícula de coordenadas (x, y) para la función f(z) = y * sin(x)
X, Y = np.meshgrid(valor_x, valor_y)

# Evaluar la función f(z) = y * sin(x) en la cuadrícula
Z = funcion_seno(X, Y)

# Visualizar la función  f(z) = y * sin(x) en un gráfico 3D
fig = plt.figure(figsize=(10,5.5))
ax = fig.add_subplot(111, projection='3d')#sistema de ejes tridimensional
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(z)')
ax.set_title('Función f(z) = y * sin(x)')
plt.show()
```

## 3. Ruido
<p align="justify">
Crear una señal (senoidal, cosenoidal, etc) y demuestra la agregación de ruido.
    


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal senoidal
amplitud = 1.0  # Amplitud de la señal
frecuencia = 5  # Frecuencia de la señal en Hz
frecuencia_muestreo = 1000  # Frecuencia de muestreo en Hz

# Genero el vector de tiempo
t = np.linspace(0, 1, int(frecuencia_muestreo))

# Genero la señal senoidal pura
senal = amplitud * np.sin(2 * np.pi * frecuencia * t)

# Genero ruido aleatorio
amplitud_ruido = 0.1  # Amplitud del ruido
ruido = amplitud_ruido * np.random.randn(len(t))  # Ruido aleatorio con distribución normal

# Agreg0 ruido a la señal
senal_ruido = senal + ruido

# Graficar la señal senoidal pura
plt.figure(figsize=(6, 4))
plt.plot(t, senal, color='red')
plt.title('Señal Senoidal sin Ruido')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()

# Graficar la señal con ruido
plt.figure(figsize=(6, 4))
plt.plot(t, senal_ruido, color='black')
plt.title('Señal Senoidal con Ruido')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

El ruido rosa, también conocido como ruido 1/f, es un tipo de ruido aleatorio que exhibe una densidad espectral de potencia que decae inversamente con la frecuencia. Esto significa que la potencia del ruido rosa en una banda de frecuencia dada es proporcional al inverso de la frecuencia en esa banda. En otras palabras, el ruido rosa tiene más energía en las frecuencias bajas y menos en las frecuencias altas.

La diferencia principal entre el ruido rosa y el ruido gaussiano radica en cómo se distribuye la energía del ruido en el dominio de la frecuencia. Mientras que el ruido gaussiano tiene una distribución de energía constante en todas las frecuencias, el ruido rosa tiene más energía concentrada en las frecuencias bajas. para demostrar la diferencia entre el ruido gaussiano y el ruido rosa, podemos generar ambos tipos de ruido y trazar su espectro de potencia, que mostrará cómo se distribuye la energía en función de la frecuencia.
